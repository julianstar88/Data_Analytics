```python
import pandas as pd
```

```python
current_city = 'LosAngeles'
```

```python
listings = pd.read_csv(f'../data/cities/processed/listings_{current_city}_processed.csv')
reviews = pd.read_csv(f'../data/cities/reviews_{current_city}_en_predictions_added.csv.gz', compression='gzip')
```

```python
listings.rename(columns={'listing_id':'id'}, inplace=True)
```

```python
listings.head(2)
```

```python
reviews.head(2)
```

```python
#reviews[reviews.listing_id == 50904]
```

```python
avg_scores_per_listing_id = reviews.groupby('listing_id')[['xgb_predict', 'rf_predict']].mean()
avg_scores_per_listing_id.rename(columns={'xgb_predict':'xgb_predict_avg', 'rf_predict':'rf_predict_avg'}, inplace=True)
```

```python
avg_scores_per_listing_id.head()
```

```python
listings_finished =  pd.merge(listings, avg_scores_per_listing_id, left_on='id', right_on='listing_id', how='inner')
```

```python
listings_finished.head()
```

```python
listings_finished.to_csv(f'../data/cities/finished/listings_{current_city}_finished.csv.gz', compression='gzip')
```
